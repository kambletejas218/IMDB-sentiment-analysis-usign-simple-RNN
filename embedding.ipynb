```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = ['the glass of milk',
        'the glass of juice',
        'the cup of tea',
        'I am a good boy',
        'I am a good developer',
        'understand the meaning of words',
        'your videos are good',]
sent
```

```python
## Define the vocabulary size
voc_size = 10000
```

```python
## One-Hot representaion
one_hot_repre = [one_hot(words,voc_size)for words in sent]
one_hot_repre
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np 
```

```python
sent_legnth = 8
embedded_docs = pad_sequences(one_hot_repre,padding='pre',maxlen=sent_legnth)
print(embedded_docs)
```

```python
## Feautre Represention 
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length =sent_legnth))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
